```python
import numpy as np
import matplotlib.pyplot as plt 
import math
import random


events = [
(+1,0),
(-1,0),
(0,+1),
(0,-1)]

def Sim_func(size,maxtime,x1,x2):
    ns = 0
    time = 0
    no_events = np.arange(0,4,1)
    t = np.zeros(size)
    v = np.zeros((size, 2))
    v[0] = np.zeros((size, 2))[0] + (y1,z1)
 
    while time <= maxtime:
        if maxtime/3<time<2*maxtime/3:
            x=x1
        else:
            x=x2
        

        
        
        rates = np.array([By+(((x/k1)**h)/(1+((x/k1)**h)))*b1,ay*v[(ns, 0)],
                          Bz+ (((x/k1)**h)/(1+((x/k1)**h)))*(1/(1+(v[(ns, 0)]/k3)**h))*b2,
                          az*v[(ns, 1)]])

    
        total_rate  =  sum(rates)

        dt = -math.log(1 - random.uniform(0.0, 1.0)) / total_rate 

        choose = np.random.choice(no_events, size=1, p=rates/total_rate)[0]
        time = time +dt
        t[ns+1]= t[ns]+dt
        v[ns+1] = v[ns]+events[choose]
        ns = ns + 1
        if ns == size-1:
            break
        if np.all((v[ns]==0)==True):
            break

    return t[0:ns], v[0:ns]
```

```python
import numpy as np
import matplotlib.pyplot as plt 
import math
import random

b1 =3
b2 =3
ay =0.01
az =0.01
k1 =1
k2 =1
k3 =10
By =1
Bz =1
h  =0.1
y1 =10
z1 =10
events = [
(+1,0),
(-1,0),
(0,+1),
(0,-1)]

def average1(size,no_steps,x1,x2):
    time = 0
    no_events = np.arange(0,4,1)
    t = np.zeros(size)
    v = np.zeros((size, 2))
    v[0] = np.zeros((size, 2))[0] + (y1,z1)
    
    
    for i in range(no_steps-1):
 

        if no_steps/3<i<2*no_steps/3:
            x=x1
        else:
            x=x2
        

        
        
        rates = np.array([By+(((x/k1)**h)/(1+((x/k1)**h)))*b1,ay*v[(i, 0)],
                          Bz+ (((x/k1)**h)/(1+((x/k1)**h)))*(1/(1+(v[(i, 0)]/k3)**h))*b2,
                          az*v[(i, 1)]])

    
        total_rate  =  sum(rates)

        dt = -math.log(1 - random.uniform(0.0, 1.0)) / total_rate 

        choose = np.random.choice(no_events, size=1, p=rates/total_rate)[0]
        time = time +dt
        t[i+1]= t[i]+dt
        v[i+1] = v[i]+events[choose]

        if np.all((v[i]==0)==True):
            break

    return  t[0:no_steps], v[0:no_steps]


def average2(size,no_steps,x1,x2):
    time = 0
    no_events = np.arange(0,4,1)
    t = np.zeros(size)
    v = np.zeros((size, 2))
    v[0] = np.zeros((size, 2))[0] + (y1,z1)
 
    for i in range(no_steps-1):
        
        if no_steps/3<i<2*no_steps/3:
            x=x1
        else:
            x=x2
        

        
        
        rates = np.array([By+(1/(1+((x/k1)**h)))*b1,ay*v[(i, 0)],
                          Bz+ ((1/(1+((x/k2)**h)))*1/(1+(v[(i, 0)]/k3)**h))*b2,
                          az*v[(i, 1)]])

    
        total_rate  =  sum(rates)

        dt = -math.log(1 - random.uniform(0.0, 1.0)) / total_rate 

        choose = np.random.choice(no_events, size=1, p=rates/total_rate)[0]
        time = time +dt
        t[i+1]= t[i]+dt
        v[i+1] = v[i]+events[choose]

        if np.all((v[i]==0)==True):
            break

    return  t[0:no_steps], v[0:no_steps]
```

```python
t1 = np.zeros(100000)
v1 = np.zeros((100000, 2))
for i in range(100):
    t,v = average1(size = 1000000,no_steps = 100000,x1 = 1,x2=0)
    t1 = t1+t
    v1 = v1+v
ave_v1 =v1/100
ave_t1 =t1/100
```
