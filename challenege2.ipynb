Population converges to stationary distribution at the carrying capacity, starting at three different initial values:

```python
import numpy as np
import matplotlib.pyplot as plt

T = 3000
x = np.zeros(T)
k = 50
r = 0.004
n = np.array([100,20,50]) 
colors = ['Red', 'Green', 'blue']
for i in range(len(n)):
 x[0] = n[i]  
 for t in range(T - 1):
 
  birth   = np.random.rand() <= r*(x[t] - (x[t]**2)/(2*k))
  death   = np.random.rand() <= r*(x[t]**2)/(2*k)
  
  x[t+1] =  x[t] +(1*birth-1*death)

 plt.plot(x,color=colors[i]);

```

Sampling population at the stationary distribution gives us normal distribution with mean value = carrying capacity (50)

```python


y = np.zeros(1000)

for j in range(1000):
 for t in range(T - 1):
    
  birth   = np.random.rand() <= r*(x[t] - (x[t]**2)/(2*k))
  death   = np.random.rand() <= r*(x[t]**2)/(2*k)
 
  x[t+1] =  x[t] +(1*birth-1*death)
 y[j] = x[T-1]
 
plt.figure(figsize=(6,3));
plt.hist(y)
plt.show()
    



```
