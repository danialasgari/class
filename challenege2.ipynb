
Population converges to the quasi-stationary distribution at the carrying capacity, starting at three different initial values:

```python
import numpy as np
import matplotlib.pyplot as plt




T = 3000
x = np.zeros(T)
k = 50
r = 0.004
n = np.array([100,20,50]) 
colors = ['Red', 'Green', 'blue']
for i in range(len(n)):
 x[0] = n[i]  
 for t in range(T - 1):
 
  birth   = np.random.rand() <= r*(x[t] - (x[t]**2)/(2*k))
  death   = np.random.rand() <= r*(x[t]**2)/(2*k)
  
  x[t+1] =  x[t] +(1*birth-1*death)

 plt.plot(x,color=colors[i]);
plt.xlabel('Time');
plt.ylabel('Population size');
```

Sampling population at the quasi-stationary distribution gives us normal distribution with mean value = carrying capacity (50)

```python

y = np.zeros(10000)

for j in range(len(y)):
 for t in range(T - 1):
    
  birth   = np.random.rand() <= r*(x[t] - (x[t]**2)/(2*k))
  death   = np.random.rand() <= r*(x[t]**2)/(2*k)
 
  x[t+1] =  x[t] +(1*birth-1*death)
 y[j] = x[T-1]
 

plt.hist(y)
plt.show()
```

Mean time value to extinction increases as the initial population size increases. 

```python
T = 5000000
x = np.zeros(T)
k = 8
r = 0.015
n =  np.arange(1,16,1)
repeat  = []
nrepeat = 20
extinct = []
for i in range(len(n)):
 x[0] = n[i]  
 for j in range(nrepeat):
  for t in range(T - 1):
   if x[t] == 0:
    repeat.append(t)
    break
   else:
    birth   = np.random.rand() <= r*(x[t] - (x[t]**2)/(2*k))
    death   = np.random.rand() <= r*(x[t]**2)/(2*k)
    x[t+1]  = x[t] +(1*birth-1*death)
 if len(repeat)!=0:  
  ave = sum(repeat)/len(repeat)
  extinct.append(ave)
```

```python
plt.plot(n,extinct)
plt.xlabel('Iinitial population size');
plt.ylabel('Mean time to extinction');
```
