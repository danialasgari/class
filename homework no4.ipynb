### Question1

The Nash equilibrium is achieved when players are not incentivised to alter their strategies. Therefore, every choice of s1 and s2 which results in equality s1+s2 = 1,000,000 should be a Nash equilibrium. If any player chooses a higher value they would lose everything and of course no one is inclined to choose a lower value. As a result we have infinite equilibria along the line s1+s2 = 1,000,000.
